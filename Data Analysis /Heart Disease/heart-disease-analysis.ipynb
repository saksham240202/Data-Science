**Distribution Plot For Continuous Variables**

```python
sns.histplot(df['RestingBP'] ,kde=True)
```

```python
sns.histplot(df['Cholesterol'] ,kde=True)
```

```python
sns.histplot(df['MaxHR'] ,kde=True)
```

**Distribution Plot for Categorical Variables**

```python
df.groupby('Sex').size().plot(kind='pie',autopct='%0.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct='%0.1f')
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f')
```

**Violin Plot**

```python
sns.violinplot(df['Age'])
```

```python
sns.violinplot(x='Sex', y='HeartDisease',data=df)
```
